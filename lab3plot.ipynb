Draw a line in a diagram from position (1,3) to (2,10) then to (6,12) and finally to position
(18,20). Mark each point with a beautiful green colour and set line colour to red and line
style dotted.

```python
import matplotlib.pyplot as plt
import numpy as np
x=[1,2,6,18]
y=[3,10,12,20]
plt.plot(x,y,color="red",marker='^',linestyle='--',mfc='g',mec='g')

plt.grid(True)
plt.show()
```

Write a Python program to plot two or more lines on the same plot with suitable legends of
each line.

```python
age=[18,21,23,24,25,26,28]
w=[55,57,58,60,66,81,85]
h=[145,146,176,156,185,175,185]
plt.plot(age,w,color='red',marker='o',label="Age VS Weight ",linestyle=':')
plt.plot(age,h,color='b',marker='^',linestyle='-',label="Age VS Height ")
plt.title("Muliple line")
plt.xlabel("Age")
plt.ylabel("Height and weight")
plt.legend()
plt.show()
```

fig, axs = plt.subplots(2, 2, figsize=(10, 8))

```python
fig,axs=plt.subplots(2,2,figsize=(10,8))
axs[0,0].plot(age,w,color='red',marker='o',label="Age VS Weight ",linestyle=':')
axs[0,0].set_title("Age VS Weight")
axs[0,0].set_xlabel("Age")
axs[0,0].set_ylabel("Weight")

axs[0,1].plot(age,h,color='green',marker='o',label="Age VS Weight ",linestyle=':')
axs[0,1].set_title("Age VS Height")
axs[0,1].set_xlabel("Age")
axs[0,1].set_ylabel("Height ")

p=[10,14,17,15,18,13,18]
axs[1,0].bar(age,p,color='blue')
axs[1,0].set_title("Age VS Position")
axs[1,0].set_xlabel("Age")
axs[1,0].set_ylabel("Positon ")

v=[11,24,37,45,58,63,78]
axs[1,1].hist(v,bins=5,color='cyan')
axs[1,1].set_title("HISTOGRAM")
axs[1,1].set_xlabel("Value")
axs[1,1].set_ylabel("Frequency ")

plt.tight_layout()
plt.show()
```
