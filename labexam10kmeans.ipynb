Write a program to implement K Means Clustering on a random dataset.

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
data,_=make_blobs(n_samples=300)
model=KMeans(n_clusters=3,random_state=24,n_init='auto')
model.fit(data)
labels=model.labels_
set(labels)
centrodis=model.cluster_centers_
plt.scatter(data[:,0],data[:,1],c=labels,cmap='viridis')
plt.scatter(centrodis[:,0],centrodis[:,1],c='red')
#plt.legend([centrodis])
plt.title("Kmeans clustering")
plt.show
```

```python
new_data=[[2,-4],[6,6],[2,5]]
result=model.predict(new_data)
for i, point in enumerate(new_data):
    print(f"The point {point} belongs to cluster {result[i]}.")
```
