Write a program to implement Linear Regression on a random dataset.


```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score
import matplotlib.pyplot as plt

```

```python
np.random.seed(42)
X=2*np.random.rand(100,1)
y=4+3*X+np.random.rand(100,1)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.2,random_state=42)
```

```python
model=LinearRegression()
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
se=mean_squared_error(y_test,y_pred)
print("mean_squared_error :",se)
r2=r2_score(y_test,y_pred)
print("r2_score :",r2)
```

```python
plt.scatter(X,y,marker='*',color='red',label='Actual')
plt.plot(X_test,y_pred,linestyle=':',color='cyan',label='Predicted')
plt.title("X vs Y")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```
